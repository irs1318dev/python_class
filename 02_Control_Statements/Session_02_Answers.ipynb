### 2. Use a while loop to calculate the factorial of 13, 18, and 1318.

```python
idx = 13
result = 1
while idx > 0:
    result *= idx
    idx -= 1
result
```

### 3. Create a list containing all integers between 1 and 1318 that are multiples of 13 or 18

```python
result = []
for idx in range(1319):
    if (idx % 13 == 0) or (idx % 18 == 0):
        result.append(idx)
str(result)
```

### 4. Use a for loop and range() to reverse the letters in "Issaquah Robotics Society". The += operator might also come in handy.

```python
IRS_str = 'Issaquah Robotics Society'
result = ''
for idx in range(len(IRS_str) - 1, -1, -1):
    result += IRS_str[idx]
result
```

### 5. Use a for loop and to print out all the numbers between 0 and 100, except for numbers that are multiples of 9 (use the continue statement and % operator)

```python
for idx in range(0, 101):
    if idx % 9 == 0:
        continue
    print(idx, end=', ')
```

### Load the JSON Files

```python
import json
sched_file = open('sched_turing_2018.json')  # JSON file must be in same folder as jupyter notebook file
sched_data = json.load(sched_file)
sched_file.close()

rank_file = open('rank_turing_2018.json')  # JSON file must be in same folder as jupyter notebook file
rank_data = json.load(rank_file)
rank_file.close()
```

### 1. Analyze the JSON schedule to find all matches with a specific team

```python
sched_data_list = sched_data['Schedule']
team_num = 1318
print('Matches for team:', team_num)
for match in sched_data_list:
    match_number = match['matchNumber']
    for team in match['teams']:
        if team['teamNumber'] == team_num:
            print(match_number)
```

### 2. Analyze the JSON schedule to find a match where two specified teams play each other.

```python
sched_data_list = sched_data['Schedule']
team1 = 1318
team2 = 2471
common_match = None
for mtch in sched_data_list:
    tm1_in_match = False
    tm2_in_match = False
    for tm in mtch['teams']:
        if tm['teamNumber'] == team1:
            tm1_in_match = True
        elif tm['teamNumber'] == team2:
            tm2_in_match = True
    if tm1_in_match and tm2_in_match:
        common_match = mtch['matchNumber']
        print(mtch['matchNumber'])
if not common_match:
    print(team1, 'and', team2, 'did not play each other.')
```
