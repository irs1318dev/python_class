```python
sql = sqlalchemy.text("""
SELECT match, alliance, "autoScaleOwnershipSec" FROM scores;
""")

scores = pd.read_sql(sql, conn)
scores
```

```python
sql = sqlalchemy.text("""
SELECT schedule.team, scores.match, scores.alliance,
       scores."autoScaleOwnershipSec"
  FROM scores INNER JOIN schedule
    ON schedule.match = scores.match AND
       schedule.alliance = scores.alliance;
""")
scores = pd.read_sql(sql, conn)
scores.shape
```

```python
sql = sqlalchemy.text("""
SELECT schedule.team, scores.match, scores.alliance,
       scores."autoScaleOwnershipSec"
  FROM scores INNER JOIN schedule
    ON schedule.match = scores.match;
""")
scores = pd.read_sql(sql, conn)
scores.shape
```

```python
sql = sqlalchemy.text("""
SELECT schedule.team, SUM(scores."autoScaleOwnershipSec") AS "autoScaleOwnershipSec"
  FROM scores INNER JOIN schedule
    ON schedule.match = scores.match
 GROUP BY schedule.team
 ORDER BY "autoScaleOwnershipSec" DESC;
""")
scores = pd.read_sql(sql, conn)
scores
```

```python
sql = sqlalchemy.text("""
SELECT count("alliance") AS color_aligned_matches
FROM scores
WHERE alliance = 'Blue' AND "switchRightNearColor" = "scaleNearColor";
""")
scores = pd.read_sql(sql, conn)
scores
```

```python
sql = sqlalchemy.text("""
SELECT avg("totalPoints")
FROM scores
GROUP BY "alliance";
""")
scores = pd.read_sql(sql, conn)
scores
```

```python
sql = sqlalchemy.text("""
SELECT teams."stateProv", avg(scores."totalPoints")
  FROM scores INNER JOIN (schedule INNER JOIN teams
    ON teams.team = to_number(schedule.team, '9999'))
	ON schedule.match = scores.match AND schedule.alliance = scores.alliance
 GROUP BY teams."stateProv";
""")
scores = pd.read_sql(sql, conn)
scores
```

```python
sql = sqlalchemy.text("""
SELECT count(team) AS numTeams, city FROM teams
GROUP BY city
ORDER BY numTeams DESC;
""")
num_tms = pd.read_sql(sql, conn)
num_tms.head()
```

```python
sql = sqlalchemy.text("""
SELECT schedule.team, avg(scores."vaultForceTotal" + scores."vaultForcePlayed" + scores."vaultLevitateTotal") AS total_cubes
  FROM scores INNER JOIN schedule
    ON schedule.match = scores.match AND schedule."alliance" = scores."alliance"
 GROUP BY schedule.team
 ORDER BY total_cubes DESC;
""")
t_cubes = pd.read_sql(sql, conn)
t_cubes.head()
```
